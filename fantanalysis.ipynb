```python
import pandas as pd

# Display options
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

```python
# Load Data
def loadData(file):
    return pd.read_excel(file, header=1)
```

```python
# Save Data to csv (single files)
def save_csv(file, year):
    file.to_csv(f"db/stat_{year}.csv")
```

```python
# Load xlsx
df21 = loadData("Statistiche_Fantacalcio_2021-22.xlsx")
df20 = loadData("Statistiche_Fantacalcio_2020-21.xlsx")
df19 = loadData("Statistiche_Fantacalcio_2019-20.xlsx")
df18 = loadData("Statistiche_Fantacalcio_2018-19.xlsx")
df17 = loadData("Statistiche_Fantacalcio_2017-18.xlsx")
```

```python
# separate ruoli
def sepRuoli(df, ruolo):
    return df.loc[df['R'] == ruolo]
```
